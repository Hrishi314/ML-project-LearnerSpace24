```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
cartoon = cv2.imread("/content/drive/MyDrive/ML Assignment Data/Assignment  3/images/cartoon.bmp")
cartoon_y = cv2.imread("/content/drive/MyDrive/ML Assignment Data/Assignment  3/images/cartoon_y.jpg")
girl = cv2.imread("/content/drive/MyDrive/ML Assignment Data/Assignment  3/images/girl.png") #grayscale image
girl_y = cv2.imread("/content/drive/MyDrive/ML Assignment Data/Assignment  3/images/girl_y.jpg")
fingerprint = cv2.imread("/content/drive/MyDrive/ML Assignment Data/Assignment  3/images/fingerprint.jpg")
fingerprint_y = cv2.imread("/content/drive/MyDrive/ML Assignment Data/Assignment  3/images/fingerprint_y.jpg")
```

```python
fig, axs = plt.subplots(3,2,figsize=(8,12))
axs[0,0].imshow(cv2.cvtColor(cartoon,cv2.COLOR_BGR2RGB))
axs[0,1].imshow(cartoon_y)
axs[0,0].axis("off")
axs[0,1].axis("off")
axs[0,0].set_title("Raw")
axs[0,1].set_title("Masked")
axs[1,0].imshow(girl)
axs[1,1].imshow(girl_y)
axs[1,0].axis("off")
axs[1,1].axis("off")
axs[1,0].set_title("Raw")
axs[1,1].set_title("Edges")
axs[2,0].imshow(cv2.cvtColor(fingerprint,cv2.COLOR_BGR2RGB))
axs[2,1].imshow(fingerprint_y)
axs[2,0].axis("off")
axs[2,1].axis("off")
axs[2,0].set_title("Raw")
axs[2,1].set_title("Binary")
plt.show()
```

```python
#START CODE HERE (Do all your processing here)
cartoon_hsv = cv2.cvtColor(cartoon_y,cv2.COLOR_RGB2HSV)
lower_bound_hsv = np.array([35,100,100])
upper_bound_hsv = np.array([85,255,255])
cartoon_y_mentee = cv2.inRange(cartoon_hsv,lower_bound_hsv,upper_bound_hsv) #(in RGB format only)
cartoon_y_mentee = cartoon_y_mentee/255.0
#END CODE HERE (the final processed image should be cartoon_y_mentee)
```

```python
#START CODE HERE (Do all your processing here)
gaussian_filter = cv2.GaussianBlur(girl_y, (1,1), 0) #Applying Gaussian Filter for better smoothing
girl_y_mentee = cv2.Canny(gaussian_filter, 100, 200)

#END CODE HERE (the final processed image should be girl_y_mentee)
```

```python
#START CODE HERE (Do all your processing here)

_,fingerprint_y_mentee = cv2.threshold(fingerprint_y,127,255,cv2.THRESH_BINARY) #(in Binary format only i.e. pixel intensity = 0 or 255)

#END CODE HERE (the final processed image should be fingerprint_y_mentee)
```

```python
fig, axs1 = plt.subplots(3,2,figsize=(8,12))
axs1[0,0].imshow(cartoon_y_mentee)
axs1[0,1].imshow(cartoon_y)
axs1[0,0].axis("off")
axs1[0,1].axis("off")
axs1[0,0].set_title("Masked_Mentee")
axs1[0,1].set_title("Masked")
axs1[1,0].imshow(girl_y_mentee,cmap="gray")
axs1[1,1].imshow(girl_y)
axs1[1,0].axis("off")
axs1[1,1].axis("off")
axs1[1,0].set_title("Edges_Mentee")
axs1[1,1].set_title("Edges")
axs1[2,0].imshow(fingerprint_y_mentee,cmap="gray")
axs1[2,1].imshow(fingerprint_y)
axs1[2,0].axis("off")
axs1[2,1].axis("off")
axs1[2,0].set_title("Binary_Mentee")
axs1[2,1].set_title("Binary")
plt.show()
```
